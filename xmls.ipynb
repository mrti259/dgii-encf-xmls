```python
import os
from pathlib import Path

base_path = Path(".")
files = os.listdir(base_path)
list(files)
```

```python
pattern = "ImpuestoAdicional"
results = []

def find_pattern(filename, pattern):
  with open(base_path / filename) as file:
    for line in file.readlines():
      if pattern in line:
        return filename

for filename in files:
  if not filename.endswith(".xml"):
    continue
  if not find_pattern(filename, pattern):
    continue
  results.append(filename)

results
```

```python
with open(base_path / results[0]) as file:
  print(file.read())
```
